```python
data_latin = {
    'letter': ['E', 'S', 'I', 'A', 'R', 'N', 'T', 'O', 'L', 'C', 'D', 'U', 'G', 'P', 'M', 'H', 'B', 'Y', 'F', 'V', 'K', 'W', 'Z', 'X', 'J', 'Q'],
    'texts':['12.7', '6.3', '7.0', '8.2', '6.0', '6.7', '9.1', '7.5', '4.0', '2.8', '4.3', '2.8', '2.0', '1.9', '2.4', '6.1', '1.5', '2.0', '2.2', \
             '0.98', '0.77', '2.4', '0.074', '0.15', '0.15', '0.095']

}

# Теперь переме

df_latin = pd.DataFrame(data_latin)
df_latin['texts'] = pd.to_numeric(df_latin['texts'])
df_latin = df_latin.sort_values(by='texts', ascending=False).reset_index(drop=True)
df_latin
```

```python
left_latin = {'letter': [], 'freq': []}
right_latin = {'letter': [], 'freq': []}

total_freq = df_latin['texts'].sum()
target_freq = total_freq / 2

current_left_latin = 0
current_right_latin = 0

for i, letter in df_latin.iterrows():
    if current_left_latin < current_right_latin:
        left_latin['letter'].append(letter['letter'])
        left_latin['freq'].append(df_latin['texts'][i])
        current_left_latin += df_latin['texts'][i]
    else:
        right_latin['letter'].append(letter['letter'])
        right_latin['freq'].append(df_latin['texts'][i])
        current_right_latin += df_latin['texts'][i]

print('left_latin','\n', pd.DataFrame(left_latin), '\n', current_left_latin)
print('right_latin', '\n', pd.DataFrame(right_latin), '\n', current_right_latin)
```

```python
left_latin_df = pd.DataFrame(left_latin)
left_latin_df = left_latin_df[:-1]
left_latin_df
```
